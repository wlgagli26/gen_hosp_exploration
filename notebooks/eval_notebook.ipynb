# Evaluation Notebook

```python
import numpy as np
import pandas as pd
from sklearn.metrics import (mean_absolute_error, root_mean_squared_error, r2_score, 
                             accuracy_score, classification_report, confusion_matrix)
from pathlib import Path
```

```python
# Load predictions

rf_preds = pd.read_parquet('../preds/rf_preds.parquet')[['Hospital overall rating', 'predicted']]
xgb_preds = pd.read_parquet('../preds/xgb_preds.parquet')[['Hospital overall rating', 'predicted']]

```

```python
# Evaluation metrics on continuous predictions:

preds_to_eval = [rf_preds, xgb_preds]
model_names = ["RF", "XGB"]
target = 'Hospital overall rating'

for i, model in enumerate(preds_to_eval):
    model = model[[target, 'predicted']].dropna().copy()
    y_true, y_pred = model[target], model['predicted']
    rmse = root_mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f'{model_names[i]} Model\nRMSE: {rmse}\nMAE: {mae}\nR2: {r2}\n')
```

```python
#Evaluation metrics on rounded/classified predictions

for i, model in enumerate(preds_to_eval):
    model = model[[target, 'predicted']].dropna().copy()
    y_true, y_pred = model[target], model['predicted'].round().clip(1,5).astype(int)
    rmse = root_mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    acc = accuracy_score(y_true, y_pred)
    
    print(f'{model_names[i]} Model\nRMSE: {rmse}\nMAE: {mae}\nR2: {r2}')
    print(f"Accuracy: {acc}")
    print(confusion_matrix(y_true, y_pred))
    print(classification_report(y_true, y_pred))
    print()
```
